```python
import cobra.io
```

```python
model = cobra.io.load_matlab_model('iGEL601.mat')
```

```python
# Copies metabolite names (already kegg ids) into annotation kegg id
unannotated = []
for met in model.metabolites:
    if met.id[0] == 'C' and met.id[1].isdigit():
        met.annotation["kegg.compound"] = met.id.rstrip('[]')
    else:
        unannotated.append(met.id)
        
```

```python
for met in model.metabolites:
    print met.annotation
```

```python
# Copies reaction names (already kegg ids) into annotation kegg id
unannotated = []
for rxn in model.reactions:
    if rxn.id[0] == 'R' and rxn.id[1].isdigit():
        rxn.annotation["kegg.reaction"] = rxn.id
    else:
        unannotated.append(rxn.id)
```

```python
for rxn in model.reactions:
    print rxn.annotation
```

```python
for gene in model.genes:
    gene.annotation['kegg.genes'] = gene.id
```

```python
#Output
cobra.io.write_sbml_model(model, 'iGEL601.xml')
cobra.io.save_json_model(model, 'iGEL601.json')
cobra.io.save_matlab_model(model, 'iGEL601.mat')
```
